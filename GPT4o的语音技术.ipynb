


autoencoder的时候 在中间再加入一个语言模型 得到对应的接龙。编码器可以是code也可以得到对应的token，可以是两者的混合。这样就不用重复制作轮子了。

这个语音模型也可以通过youtube的视频进行预训练，然后进行finetune和RLHF


微调的时候一般只需要进行说话人少量的数据就可以改变GPT的音色。可能是因为pretrain了，也可能是用了语音替换的技术



那有一个问题就是语音模型如何知道何时说话人结束说话。对于文字来说，按下发送就知道end。但是语音模型则不会这样。所以可以听力、语音自己说话和视力三个通道。如果说话人安静就继续说 我好期待。若说话人继续说，就保持安静。同时可以有视力。当说话人问到以前看到的东西，就做attention，给出对应的结果
