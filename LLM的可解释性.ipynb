## 可解释性的三个说法



透明性是能够知道具体的参数 甚至是知道训练资料和训练过程

interpretable是可说明的 是指能够说明具体的结构。就像决策树一样能够得到具体的分叉是什么意思

explainable是可解释性，就是解释为什么这么输出。比如李宏毅为什么叫大金，有个具体的原因。但是transformer是不可说明的，因为具体的结构干了什么不知道。

## 分析embedding的信息

对中间的输入进行probing 分类出词性，看出中间是什么过程。

一般来说刚开始表面的信息，之后的语法的，最后几层是语义的，逐渐抽象，但也不绝对。

但这种方式也有点问题，因为有可能一个层能包括很多信息，模型能力很强，就是知道我们人类不能了解的东西，所以分类出来的东西可能没有意义。因为模型可能这一层就是要弄什么层次信息都有的东西。

##  新的可解释性的研究方向----直接问LLM


前图是问每个token的重要程度。研究得到这个给的信息分数和模型实际的分数还是非常靠近的



针对某一个特定的神经元，越绿权重越高，直接问GPT4进行解释。然后再用一个新的GPT4根据解释得到这个文字中各个token的权重 与真实情况对比。从而得到解释可靠分数



或者用GPT4直接扮演神经元，看最终输出的变化程度，从而德奥解释可靠分数。这两种解释可靠分数是正相关的




某一个神经元的解释是在找ALL 那就生成相似句子（有很多all的句子 再次进行实验）  最后发现实际在找not all。这是一种revise的方法，能够增强可解释性。

LLM的可解释得分能到0.14(0-1越大越好)，但是人类其实也只能得到0.18.说明其实LLM是可行的，但这个方向还有很大的进步空间
